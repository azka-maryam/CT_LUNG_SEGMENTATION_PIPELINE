```python
import SimpleITK as sitk
import pandas as pd
import numpy as np
import os

# Input and output folders
input_folder = r"C:\Users\R c\Documents\CT_pipeline_Portfolio\raw_data"
output_folder = r"C:\Users\R c\Documents\CT_pipeline_Portfolio\csv_results"
os.makedirs(output_folder, exist_ok=True)

summary = []

# Loop through all .nii.gz files in the raw_data folder
for file in os.listdir(input_folder):
    if file.endswith(".nii.gz"):
        path = os.path.join(input_folder, file)
        img = sitk.ReadImage(path)
        arr = sitk.GetArrayFromImage(img)

        stats = {
            "filename": file,
            "size_x": img.GetSize()[0],
            "size_y": img.GetSize()[1],
            "size_z": img.GetSize()[2],
            "spacing_x": img.GetSpacing()[0],
            "spacing_y": img.GetSpacing()[1],
            "spacing_z": img.GetSpacing()[2],
            "origin_x": img.GetOrigin()[0],
            "origin_y": img.GetOrigin()[1],
            "origin_z": img.GetOrigin()[2],
            "mean_intensity": float(np.mean(arr)),
            "min_intensity": float(np.min(arr)),
            "max_intensity": float(np.max(arr))
        }

        summary.append(stats)

# Create a DataFrame and save to CSV
df = pd.DataFrame(summary)
output_path = os.path.join(output_folder, "raw_data_stats.csv")
df.to_csv(output_path, index=False)

print("✅ Stats saved to:", output_path)
print(df)
```
